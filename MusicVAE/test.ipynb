```python
def swap_dict(i,j, dict_list):
    dict_list[i], dict_list[j] = dict_list[j], dict_list[i]
    return dict_list
```

```python
dict_list = swap_dict(0,1,dict_list)
print(dict_list)
```

```python
test = np.array(dict_list[0][1])
noise = np.random.normal(0,0.00001,512)
test_noise = test + noise
norm = np.linalg.norm(test_noise - test)
print(norm)
```
