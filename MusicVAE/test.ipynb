```python
import numpy as np
def slerp(p0, p1, t):
  """Spherical linear interpolation."""
  omega = np.arccos(np.dot(np.squeeze(p0/np.linalg.norm(p0)), np.squeeze(p1/np.linalg.norm(p1))))
  so = np.sin(omega)
  return np.sin((1.0-t)*omega) / so * p0 + np.sin(t*omega)/so * p1
```

```python
Z_SIZE = 512
num_bars = 32

z1 = np.random.normal(size=[Z_SIZE])
z2 = np.random.normal(size=[Z_SIZE])
z = np.array([slerp(z1, z2, t)
              for t in np.linspace(0, 1, num_bars)])
print(type(z))
```

```python
def swap_dict(i,j, dict_list):
    dict_list[i], dict_list[j] = dict_list[j], dict_list[i]
    return dict_list
```

```python
dict_list = swap_dict(0,1,dict_list)
print(dict_list)
```

```python
test = np.array(dict_list[0][1])
noise = np.random.normal(0,0.00001,512)
test_noise = test + noise
norm = np.linalg.norm(test_noise - test)
print(norm)
```
